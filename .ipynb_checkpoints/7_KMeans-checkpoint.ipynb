# 7. KMeans

```python
# Load the Dataset
import pandas as pd

df = pd.read_csv("ml_customer_data.csv")
X = df[['age', 'salary']]
y_true = df['purchased']  # Actual labels, for optional evaluation
```

```python
# Elbow Method: Find optimal number of clusters
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
k_range = range(1, 10)

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")
plt.grid(True)
plt.show()
```

```python
# Fit KMeans (assume 2 for comparison to 'purchased')
model = KMeans(n_clusters=2, random_state=42)
df['cluster'] = model.fit_predict(X)
```

```python
# Visualize Clusters
plt.figure(figsize=(8, 6))
plt.scatter(X['age'], X['salary'], c=df['cluster'], cmap='viridis', s=50)
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("K-Means Clustering Results")
plt.legend()
plt.show()
```

```python
# Cluster Sizes
print("Cluster Distribution:")
print(df['cluster'].value_counts())
```

```python
# Silhouette Score (measures how well-separated the clusters are)
from sklearn.metrics import silhouette_score

score = silhouette_score(X, df['cluster'])
print("Silhouette Score:", round(score, 3))
```

```python
# Compare to actual labels
from sklearn.metrics import confusion_matrix

print("Confusion Matrix (Actual vs Cluster):")
print(confusion_matrix(y_true, df['cluster']))
```

## Model Summary: K-Means Clustering on Customer Purchase Data

This section summarizes the results and interpretation of applying **K-Means Clustering** (unsupervised learning) to group customers based on age and salary. The model attempts to discover natural groupings in the data without access to the actual labels.

**Cluster Count:**  
2 (chosen manually to match the binary nature of the `purchased` variable)

**Cluster Sizes:**  
- Cluster 0: 283 points  
- Cluster 1: 217 points  

This shows a reasonably balanced clustering outcome with no extreme imbalance.

---

**Silhouette Score:**  
0.622  
The silhouette score ranges from -1 to 1. A value closer to 1 means clusters are well separated.  
- A score of 0.622 indicates **good clustering structure**, with distinct boundaries between groups.

---

**Visual Interpretation:**  
- The 2D scatter plot with centroids (red Xs) shows clear separation between high-salary and low-salary groups.
- Clusters are influenced more by **salary** than age.
- Cluster 1 appears to capture the **higher-salary** segment, while Cluster 0 includes **lower-salary** customers.

---

**Confusion Matrix (vs. Actual Labels):**  
[[226  76]  
 [ 57 141]]

| Actual Purchased | Cluster 0 | Cluster 1 |
|------------------|-----------|-----------|
| 0 (Not Purchased) | 226       | 76        |
| 1 (Purchased)     | 57        | 141       |

Although K-Means does not use labels, this comparison shows:
- Cluster 1 aligns well with actual buyers (141 out of 198)
- Cluster 0 mostly aligns with non-buyers (226 out of 302)
- **This suggests that clustering reflects real behavior reasonably well**

---

### Final Conclusion

K-Means successfully grouped customers into two meaningful clusters based on age and salary, with a **good silhouette score (0.622)** and a visual split largely driven by salary level. When compared to actual purchase labels, the clustering aligns well, especially for identifying high-value (purchasing) segments. This demonstrates that K-Means can serve as a **useful exploratory tool** for segmenting customers even without labeled data.
